```python
#Import library opencv
import cv2

#Baca aloritma HAAR face Cascades
algoritma_wajah = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
#Baca algoritma HAAR eye cascades 
algoritma_mata = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')
#Baca algoritma HAAR mouth cascades
algoritma_mulut = cv2.CascadeClassifier('Mouth.xml')

# Buka webcam atau video capture.
web_cam = cv2.VideoCapture(0)

# Metode untuk menggambar batas label di sekitar fitur yang terdeteksi
def gambar_frame(img, classifier, scaleFactor, minNeighbors, color, text):
    # Mengubah gambar ke gray-scale
    grayscale_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    # Mendeteksi fitur dalam gambar skala abu-abu, mengembalikan koordinat, lebar dan tinggi fitur
    fitur = classifier.detectMultiScale(grayscale_image, scaleFactor, minNeighbors)
    coords = []
    # Menggambar persegi panjang di sekitar fitur dan memberi label
    for (x, y, w, h) in fitur:
        cv2.rectangle(img, (x,y), (x+w, y+h), color, 2)
        #Menambah text pada label wajah
        cv2.putText(img, text, (x, y-4),cv2.FONT_HERSHEY_TRIPLEX, 0.8, color, 1, cv2.LINE_AA)
        coords = [x, y, w, h]
    return coords

# Metode untuk mendeteksi fitur
def detect(img, algoritma_wajah, algoritma_mata, algoritma_mulut):
    #Membuat variabel warna untuk label
    color = {"hijau":(79, 235, 52), "merah":(0,0,255),"biru":(252, 3, 3)}
    #Memberi warna pada label deteksi wajah
    coords = gambar_frame(img, algoritma_wajah, 1.1, 10, color['hijau'], "Wajah")
    # Jika fitur terdeteksi, metode gambar_frame akan mengembalikan koordinat x,y dan lebar dan tinggi persegi panjang lainnya, panjang coords adalah 0
    if len(coords)==4:
        #Memperbarui range yang diinginkan dengan memotong gambar
        roi_img = img[coords[1]:coords[1]+coords[3], coords[0]:coords[0]+coords[2]]
        # Melewati roi, classifier, scaling factor, Minimum neighbours,warna,dan teks label
        # Memberi warna pada label deteksi mata
        coords = gambar_frame(roi_img, algoritma_mata, 1.1, 12, color['merah'], "Mata")
        # Memberi wana pada label deteksi mulut
        coords = gambar_frame(roi_img, algoritma_mulut, 1.1, 20, color['biru'], "Mulut")
    #Mengembalikan gambar 
    return img

while True:
    # Membaca gambar dari video web-cam
    _, img = web_cam.read()
    # Metode return algoritma yang telah terdefinisi di atas
    img = detect(img, algoritma_wajah, algoritma_mata, algoritma_mulut)
    # Tampilkan output videocam (videocapture)
    cv2.imshow("face detection", img)
    # Tekan esc untuk keluar 
    if cv2.waitKey(1)==27:
        break
        
# realese video webcam 
web_cam.release()
# menutup jendela output
cv2.destroyAllWindows()
```
